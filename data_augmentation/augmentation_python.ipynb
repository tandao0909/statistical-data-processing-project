```python
from sklearn.utils import resample
import pandas as pd

# cột chứa nhãn (target) và các đặc trưng (features)
target_column = "Diabetes_012"
X = data.drop(columns=[target_column])
y = data[target_column]

# kiểm tra phân phối nhãn
print("Phân phối ban đầu:")
print(y.value_counts())

# 1. Undersampling
undersampled_data = pd.DataFrame()
min_class_count = y.value_counts().min()

for label in y.unique():
    subset = data[data[target_column] == label]
    # ;lấy mẫu ngẫu nhiên số lượng bằng `min_class_count`
    undersampled_subset = resample(subset, replace=False, n_samples=min_class_count, random_state=42)
    undersampled_data = pd.concat([undersampled_data, undersampled_subset])

undersampled_data.to_csv("undersampled_data.csv", index=False)
print("\nĐã lưu dữ liệu undersampled vào 'undersampled_data.csv'")

# 2. Oversampling (tăng 2 nhãn ít lên bằng 80% lượng nhãn nhiều nhất)
oversampled_data = pd.DataFrame()
max_class_count = y.value_counts().max()
oversample_limit = int(max_class_count * 0.8)  # tăng đến 80% của nhãn nhiều nhất

for label in y.unique():
    subset = data[data[target_column] == label]
    if y.value_counts()[label] < oversample_limit:
        # tăng nhãn ít (dưới giới hạn) lên bằng `oversample_limit`
        oversampled_subset = resample(subset, replace=True, n_samples=oversample_limit, random_state=42)
    else:
        # giữ nguyên nhãn đã nhiều
        oversampled_subset = subset
    oversampled_data = pd.concat([oversampled_data, oversampled_subset])

oversampled_data.to_csv("oversampled_data.csv", index=False)
print("\nĐã lưu dữ liệu oversampled vào 'oversampled_data.csv'")

# kiểm tra phân phối sau cân bằng
print("\nPhân phối sau khi undersampling:")
print(undersampled_data[target_column].value_counts())

print("\nPhân phối sau khi oversampling:")
print(oversampled_data[target_column].value_counts())
```

```python
import pandas as pd
import numpy as np
from imblearn.over_sampling import SMOTE

# cột định lượng và định tính
quantitative_columns = ["BMI", "MentHlth", "PhysHlth"]
categorical_columns = data.columns.difference(quantitative_columns + ["Diabetes_012"])

# dữ liệu ban đầu
target_column = "Diabetes_012"
X = data[quantitative_columns]
y = data[target_column]

# phân phối nhãn ban đầu
label_counts = y.value_counts()
majority_count = label_counts[0.0]
augmentation_target = int(majority_count * 0.8)

# **1. Tăng dữ liệu bằng SMOTE trên toàn bộ dữ liệu định lượng**
smote = SMOTE(random_state=42)
X_smote, y_smote = smote.fit_resample(X, y)

# chuyển đổi sang DataFrame và Series
X_smote = pd.DataFrame(X_smote, columns=quantitative_columns)
y_smote = pd.Series(y_smote, name=target_column)

# chỉ giữ lại số lượng cần thiết cho các nhãn 1.0 và 2.0
X_class_2_smote = X_smote[y_smote == 2.0].iloc[:augmentation_target]
y_class_2_smote = pd.Series([2.0] * len(X_class_2_smote), name=target_column)

X_class_1_smote = X_smote[y_smote == 1.0].iloc[:augmentation_target]
y_class_1_smote = pd.Series([1.0] * len(X_class_1_smote), name=target_column)

# **2. Kết hợp dữ liệu định tính (Augmentation)**
# lấy mẫu ngẫu nhiên từ các cột định tính
def augment_categorical_data(original_data, target_count):
    augmented_samples = []
    for _ in range(target_count - len(original_data)):
        sample = original_data.sample(n=1, replace=True, random_state=np.random.randint(1000))
        augmented_samples.append(sample)

    augmented_data = pd.concat(augmented_samples, ignore_index = True)
    return pd.concat([original_data, augmented_data], ignore_index = True)

data_class_2 = data[data[target_column] == 2.0]
categorical_class_2 = augment_categorical_data(data_class_2[categorical_columns], augmentation_target)

data_class_1 = data[data[target_column] == 1.0]
categorical_class_1 = augment_categorical_data(data_class_1[categorical_columns], augmentation_target)

# **3. Kết hợp SMOTE và Augmentation**
class_2_final = pd.concat([X_class_2_smote.reset_index(drop=True), categorical_class_2.reset_index(drop=True)], axis=1)
class_2_final[target_column] = 2.0

class_1_final = pd.concat([X_class_1_smote.reset_index(drop=True), categorical_class_1.reset_index(drop=True)], axis=1)
class_1_final[target_column] = 1.0

# dữ liệu cuối cùng
data_class_0 = data[data[target_column] == 0.0]   # lớp gốc 0.0
final_data = pd.concat([data_class_0, class_2_final, class_1_final], ignore_index=True)


final_data.to_csv("balanced_data_80_percent.csv", index=False)
print("\nĐã lưu dữ liệu vào file 'balanced_data_80_percent.csv'")


print("\nPhân phối sau khi tăng cường:")
print(final_data[target_column].value_counts())
```
